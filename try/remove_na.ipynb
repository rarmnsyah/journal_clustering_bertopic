```python
dataset = pd.read_csv('../dataset/articles.csv', sep = '|')
dataset = dataset.drop(columns = ['Unnamed: 0'])
dataset
```

```python
bodied_df = dataset.query("desc != 'TIDAK ADA' or title != 'TIDAK ADA'")
bodied_df.dropna(inplace=True)
bodied_df = bodied_df.reset_index()
bodied_df
```

```python
word_count = bodied_df.desc.apply(lambda x : len(x.split()))
word_count
```

```python
max(word_count), min(word_count)
```

```python
bodied_df['word_count'] = word_count
bodied_df[bodied_df.word_count > 1000]
```

```python
df_less_1000 = bodied_df[bodied_df.word_count < 1000].reset_index()
df_less_1000
```

```python
text = df_less_1000.sample().desc.values[0]

preprocessed_text = preprocess_text(text, [remove_abs_word, remove_nbsp, remove_punctuation, keep_alpha_numeric, remove_katakunci, remove_multilang, remove_whitespace])

lang = lang_checker_langdetect(preprocessed_text)
id_text = en_to_id(preprocessed_text) if lang == 'id' else preprocessed_text

print(text, lang)
print(preprocessed_text)
print(id_text)
```
