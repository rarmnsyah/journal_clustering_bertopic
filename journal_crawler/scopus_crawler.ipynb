```python
from crossref.restful import Works, Journals
```

```python
issn_journals = {
    # 'Studies in Mycology' : '0166-0616',
    # 'Administrative Science Quarterly' : '0001-8392',
    'Cell Metabolism' : '1550-4131',
    # 'Artificial Intelligence Review' : '1573-7462',
    'Engineering Geology' : '0013-7952',
    # 'Progress in Orthodontics' : '2196-1042',
    # 'Annual Review of Astronomy and Astrophysics' : '0066-4146',
    # 'Journal of Finance' : '0022-1082'
}
```

```python
journals = Journals()
variable = ['abstract', 'title', 'DOI', 'created', 'score', 'URL']
article = {journal : [] for journal in issn_journals.keys()}
```

```python
list(issn_journals.keys())
```

```python
journal_names = list(issn_journals.keys())

for journal in journal_names:
    try:
        work = journals.works(issn_journals[journal]).filter(has_abstract = 1).select('abstract', 'title', 'DOI', 'created', 'score', 'URL')
        for i in work:
            article[journal].append(i)
            # print(i)
    except Exception as e:
        print(e, journal)
        pass
```

```python
for art in article:
    journal = art
    for i in article[art]:
        try:
            abstract = i['abstract']
        except:
            abstract = None
        try:
            title = i['title']
        except:
            title = None
        try:
            DOI = i['DOI']
        except:
            DOI = None
        try:
            created = i['created']
        except:
            created = None
        try:
            score = i['score']
        except:
            score = None
        try:
            URL = i['URL']
        except:
            URL = None

        df['journal'].append(journal)
        df['abstract'].append(abstract)
        df['title'].append(title)
        df['DOI'].append(DOI)
        df['created'].append(created)
        df['score'].append(score)
        df['URL'].append(URL)

df, len(df)
```

```python
import pandas as pd

dataset_scopus = pd.DataFrame(df)
dataset_scopus.to_csv('../dataset/data_scopus_raw.csv', index=False)
```
