```python
import pandas as pd
from sickle import Sickle
from sickle import oaiexceptions
from IPython.display import clear_output
import xmltodict
import glob
import os
import json
from pathlib import Path
import threading
from time import sleep
from random import shuffle

STOREPATH = 'oai_sickle_response'
df = pd.read_csv('all_garuda_journalweb_scrapped.csv', usecols=['jid', 'oai'])
all_jid = df['jid'].values

def touch_file(file_path):
    Path(file_path).touch()

jid_done = set()
def get_next_jid():
    global jid_done
    jid_done = set([int(fp.split('jid')[-1].split('_')[0]) for fp in glob.glob(os.path.join(STOREPATH, 'jid*_*'))])
    new_jid = list(set(all_jid).difference(jid_done))
    if len(new_jid) == 0:
        return None
    shuffle(new_jid)
    new_jid = new_jid[0]
    touch_file(os.path.join(STOREPATH, f'jid{new_jid}_onprogress'))
    return new_jid

def construct_records(link):
    s = Sickle(link)
    records = s.ListRecords(metadataPrefix='oai_dc')
    return records

def process_jid(jid):
    global jid_done
    try:
        if jid is None:
            raise Exception('All finished!')
        link = df[df['jid'] == jid]['oai'].values[0]
        records = construct_records(link)
        downloaded_record = []
        while 1:
            try:
                r = records.next()
                res = xmltodict.parse(r.raw)
                downloaded_record.append(res)
                clear_output(wait=True)
                display(f'thread: {hex(threading.get_ident())}, done: {len(jid_done)}/{len(df)}, jid: {jid}, downloaded: {len(downloaded_record)}, link: {link}')
            except StopIteration:
                break
        with open(os.path.join(STOREPATH,f'jid{jid}_len{len(downloaded_record)}_oaisickle.json'), 'w') as file:
            json.dump(downloaded_record, file)
    except Exception as e:
        display(e)
    
    os.remove(os.path.join(STOREPATH, f'jid{jid}_onprogress'))
```

```python
max_nthread = 32

tlist = []

while 1:
    for t in tlist:
        if not t.is_alive():
            tlist.remove(t)

    while len(tlist) < max_nthread: 
        current_jid = get_next_jid()
        if current_jid is not None:
            t = threading.Thread(target=process_jid, args=(current_jid, ))
            t.start()
            tlist.append(t)

    sleep(3)
```
